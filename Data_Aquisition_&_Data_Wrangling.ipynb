# To check the Outliers and remove

```python
sns.boxplot(concatenated_data['casual'])
```

Statistical Methods using IQR method

```python
df = pd.DataFrame(concatenated_data)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Outliers using IQR method:")
print(outliers_iqr)
```

```python
concatenated_data_no_outlier = concatenated_data[(concatenated_data.casual>0)&(concatenated_data.casual<4)]

sns.boxplot(concatenated_data_no_outlier ['casual'])
```
